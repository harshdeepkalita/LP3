```python
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    gd = []
    for i in range(num_iterations):
        gradient = 2 * (x + 3)  
        x = x - learning_rate * gradient  
        y = (x + 3)**2
        gd.append(x)

        print(f"Iteration {i+1}: x = {x}, y = {y}")

    return x,gd

# Parameters
starting_x = 2
learning_rate = 0.1 
num_iterations = 900 

# Call the function
minimum_x,lx = gradient_descent(starting_x, learning_rate, num_iterations)
print(f"Local minimum occurs at x = {minimum_x}")
```
